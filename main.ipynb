```python
# fill_na(df)

def fill_na(df):
    temp = df.isna().sum()
    nulNumList = temp[temp.index[temp > 0]]
    # print(nulNumList)
    ''' // 출처: https://www.kaggle.com/code/munmun2004/house-prices-for-begginers
    LotFrontage      259        Linear feet of street connected to property
    Alley           1369        Type of alley access,  NA: No alley access
    MasVnrType       872        Masonry veneer type, NA is just None
    MasVnrArea         8        Masonry veneer area in square feet, NA is just 0
    BsmtQual          37        Evaluates the height of the basement, NA is No Basement
    BsmtCond          37        Evaluates the general condition of the basement, NA is No Basement
    BsmtExposure      38        ''
    BsmtFinType1      37        ''
    BsmtFinType2      38        ''
    Electrical         1
    FireplaceQu      690
    GarageType        81
    GarageYrBlt       81        PUT SAME AS YEAR it IS BUILT
    GarageFinish      81
    GarageQual        81
    GarageCond        81
    PoolQC          1453        //just fill na
    Fence           1179        ''
    MiscFeature     1406        ''
    '''
    # LotFrontage   259
    df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

    # Alley         1369
    df['Alley'] = df['Alley'].fillna("None")

    # MasVnrType    872
    df['MasVnrType'] = df['MasVnrType'].fillna("None")
    # MasVnrArea    8
    df['MasVnrArea'] = df['MasVnrArea'].fillna(0)

    # BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2  37~38
    for BsmtEle in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
        df[BsmtEle] = df[BsmtEle].fillna("None")

    # Electrical         1 - SBrkr이 (standard) 1334개로 젤 많음. 이걸로 ㄱㄱ
    # print(df_house.groupby('Electrical')['Electrical'].count())
    # print(df_house.groupby('Electrical')['Electrical'].size())
    df['Electrical'] = df['Electrical'].fillna('SBrkr')
        
    # FireplaceQu      690
    df["FireplaceQu"] = df["FireplaceQu"].fillna("None")

    # GarageType, GarageFinish, GarageQual, GarageCond      81
    for GarEle in ["GarageType", "GarageFinish", "GarageQual", "GarageCond"]:
        df[GarEle] = df[GarEle].fillna("None")
        
    # GarageYrBlt       81
    df['GarageYrBlt'] = df.apply(lambda x: x['YearBuilt'] if np.isnan(x['GarageYrBlt']) else x['GarageYrBlt'], axis=1)
    ''' ADDDEDE DAJEINGIONEIOANIOO '''

    # PoolQC          1453
    df["PoolQC"] = df["PoolQC"].fillna("None")

    # Fence           1179
    df["Fence"] = df["Fence"].fillna("None")

    # MiscFeature     1406
    df["MiscFeature"] = df["MiscFeature"].fillna("None")

    temp = df.isna().sum()
    nulNumList = temp[temp.index[temp > 0]]
    
    '''
    MSZoning        4
    Utilities       2
    Exterior1st     1
    Exterior2nd     1
    BsmtFinSF1      1
    BsmtFinSF2      1
    BsmtUnfSF       1
    TotalBsmtSF     1
    BsmtFullBath    2
    BsmtHalfBath    2
    KitchenQual     1
    Functional      2
    GarageCars      1
    GarageArea      1
    SaleType        1
    '''
    df['MSZoning'] = df['MSZoning'].fillna("RL") # most값
    df['Utilities'] = df['Utilities'].fillna("AllPub") # most값
    
    for ExtrLabel in ['Exterior1st', 'Exterior2nd']:
        df[ExtrLabel] = df[ExtrLabel].fillna("VinylSd") # most값
    
    for BsmtLabel in ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']:
        df[BsmtLabel] = df[BsmtLabel].fillna(0)
        
    df['BsmtFullBath'] = df['BsmtFullBath'].fillna(1) # most값
    df['BsmtHalfBath'] = df['BsmtHalfBath'].fillna(0) # most값
    
    df['KitchenQual'] = df['KitchenQual'].fillna("TA") # most값
    
    df['Functional'] = df['Functional'].fillna("Typ") # most값
    
    for GarLabel in ['GarageCars', 'GarageArea']:
        df[GarLabel] = df[GarLabel].fillna(0) # most값
    
    df['SaleType'] = df['SaleType'].fillna("WD") # most값
    
    print(nulNumList)
    # for item in nulNumList.items():
        # print(df_house[item[0]])
```

```python
# add_labels(df)
# additional labels

def add_labels(df):
    # 출처: https://www.kaggle.com/code/munmun2004/house-prices-for-begginers
    df['TotalSF'] = (df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF'])
    df['YrBltAndRemod'] = df['YearBuilt'] + df['YearRemodAdd']
    df['Total_sqr_footage'] = (df['BsmtFinSF1'] + df['BsmtFinSF2'] + df['1stFlrSF'] + df['2ndFlrSF'])
    df['Total_Bathrooms'] = (df['FullBath'] + (0.5 * df['HalfBath']) + df['BsmtFullBath'] + (0.5 * df['BsmtHalfBath']))
    df['Total_porch_sf'] = (df['OpenPorchSF'] + df['3SsnPorch'] + df['EnclosedPorch'] + df['ScreenPorch'] + df['WoodDeckSF'])

    df['haspool'] = df['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
    df['has2ndfloor'] = df['2ndFlrSF'].apply(lambda x: 1 if x > 0 else 0)
    df['hasgarage'] = df['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
    df['hasbsmt'] = df['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
    df['hasfireplace'] = df['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)

    for i in range(11):
        print(list(df.columns[i*9:(i+1)*9]))
        pass
```

```python
# remove_outlier(df)
# 이상치 제거

def remove_outlier(df):
    # LotArea
    ovc = df['LotArea']
    df.drop(ovc.index[ovc>100000], axis=0, inplace=True)
    print('LotArea - dropped: %d'%len(ovc.index[ovc>100000]))
    
    # MasVnrArea
    ovc = df['MasVnrArea']
    df.drop(ovc.index[ovc>1200], axis=0, inplace=True)
    print('MasVnrArea - dropped: %d'%len(ovc.index[ovc>1200]))
    
    # OpenPorchSF
    ovc = df['OpenPorchSF']
    df.drop(ovc.index[ovc>400], axis=0, inplace=True)
    print('OpenPorchSF - dropped: %d'%len(ovc.index[ovc>400]))
    
    # LotFrontage
    ovc = df['LotFrontage']
    df.drop(ovc.index[ovc>300], axis=0, inplace=True)
    print('LotFrontage - dropped: %d'%len(ovc.index[ovc>300]))
    
    # Total_sqr_footage
    ovc = df['Total_sqr_footage']
    df.drop(ovc.index[ovc>5500], axis=0, inplace=True)
    print('Total_sqr_footage - dropped: %d'%len(ovc.index[ovc>5500]))
    
    # TotalSF
    ovc = df['TotalSF']
    df.drop(ovc.index[ovc>6000], axis=0, inplace=True)
    print('TotalSF - dropped: %d'%len(ovc.index[ovc>6000]))
    
    # TotalSF
    ovc = df['BsmtFinSF1']
    df.drop(ovc.index[ovc>2000], axis=0, inplace=True)
    print('BsmtFinSF1 - dropped: %d'%len(ovc.index[ovc>2000]))
    
    # WoodDeckSF
    ovc = df['WoodDeckSF']
    df.drop(ovc.index[ovc>750], axis=0, inplace=True)
    print('WoodDeckSF - dropped: %d'%len(ovc.index[ovc>750]))

# df_house = remove_outlier(df_house)
    
```

```python
# analysis_skew(df)
# skew 분석
# df_house['SalePrice'] = np.log2(df_house['SalePrice'])

from scipy.stats import skew

def analysis_skew(df):
    skewness = abs(df[num_columns].apply(lambda x: skew(x.dropna()))).sort_values(ascending=False)
    high_skewness = skewness[abs(skewness) > 0.75]
    idx = high_skewness.index.tolist()
    # print(high_skewness)

    for label in idx:
        # print(label)
        k = df.groupby(label)[label].count().size
        # print(label, k)
        if k < 50: # too few categories may lead to high skew
            high_skewness.drop(label, inplace=True)
            # print("removing... %s"%label)
        else:
            zero_num = df[df[label] == 0][label].count()
            if zero_num > 200: # plotting log with too many zero will lose datas
                high_skewness.drop(label, inplace=True)
            else:
                print(label, zero_num, high_skewness[label])
            # print("label: %s has value: %d \n"%(label, k))
                pass
    idx = high_skewness.index.tolist()

    # print(high_skewness)

    row = 4
    col = 3 
    fig, axs = plt.subplots(row, col, figsize = (col*3, row*4))

    for r in range(0, row):
        for c in range(0, col):
            i = r*col + c
            if i < len(idx):
                sns.regplot(x=idx[i], y=target, data=df, ax = axs[r][c])

    # sns.displot(df_house['SalePrice']
```
